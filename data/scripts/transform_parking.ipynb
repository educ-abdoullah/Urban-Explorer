```python
import pandas as pd
import geopandas as gpd
```

```python
parkings = gpd.read_file("../raw/stationnement-en-ouvrage.geojson")
voie_publique = gpd.read_file("../raw/stationnement-voie-publique-emplacements.geojson")
quartiers_gdf = gpd.read_file("../raw/quartier_paris.geojson")
```

```python
parkings.head()
```

```python
quartiers_gdf = quartiers_gdf.to_crs(voie_publique.crs)
parkings = parkings.to_crs(quartiers_gdf.crs)
parkings_quartiers = gpd.sjoin(
    parkings,
    quartiers_gdf,
    how="inner",
    predicate="within"
)
```

```python
parkings_quartiers = parkings_quartiers[["l_qu","c_qu","id","nom","geometry","nb_places"]]
```

```python
parkings_quartiers.to_parquet("../silver/silver_parkings_quartiers.parquet")
```

```python
df_parkings = pd.read_parquet("../silver/silver_parkings_quartiers.parquet")
df_voie_publique = pd.read_parquet("../silver/silver_voie_publique_quartiers.parquet")
```

```python
categories_stationnement = [
    "PAYANT MIXTE",
    "PAYANT ROTATIF",
    "GRATUIT",
    "GIG/GIC",
    "ELECTRIQUE"
]
```

```python
parking_auto = df_voie_publique[
    df_voie_publique["regpri"].isin(categories_stationnement)
].copy()

parking_auto["plarel"] = pd.to_numeric(
    parking_auto["plarel"], errors="coerce"
).fillna(0)

stationnement_auto_quartier = (
    parking_auto.groupby(["c_qu", "l_qu"], as_index=False)
    .agg(nb_places_auto=("plarel", "sum"))
)

print(stationnement_auto_quartier.head())
```

```python
df_parkings["nb_places"] = pd.to_numeric(
    df_parkings["nb_places"], errors="coerce"
).fillna(0)

parking_ouvrage_quartier = (
    df_parkings.groupby(["c_qu", "l_qu"], as_index=False)
    .agg(nb_places_ouvrage=("nb_places", "sum"))
)

stationnement_quartier = parking_ouvrage_quartier.merge(
    stationnement_auto_quartier,
    on=["c_qu", "l_qu"],
    how="outer"
).fillna(0)

stationnement_quartier["stationnement_total"] = (
    stationnement_quartier["nb_places_ouvrage"] +
    stationnement_quartier["nb_places_auto"]
)

print(stationnement_quartier.head())
```

```python
quartiers_gdf["c_qu"] = pd.to_numeric(quartiers_gdf["c_qu"])
```

```python
stationnement_quartier["c_qu"] = pd.to_numeric(stationnement_quartier["c_qu"])
```

```python
stationnement_quartier = stationnement_quartier.merge(
    quartiers_gdf[["c_qu","surface"]],
    on="c_qu",
    how="left"
)
```

```python
stationnement_quartier["places_km2"] = (
    stationnement_quartier["stationnement_total"]/
    (stationnement_quartier["surface"]/1000000)
)
```
